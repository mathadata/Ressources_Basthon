## Feature function


!!! tip Feature function  
We will start by coding the `feature` function which takes as input an image and returns its feature number.  
  
!!!
!!! question Mean of an image  
Complete the `feature` function in the following cell, which returns the mean of the pixels of the image `d`.  
`d` is a 2D array, so a list where each element is a list. `d[i]` is row $i$ of the image, and `d[i][j]` is the value of the pixel at column $j$ of that row.  
!!!

```python
def feature(d):
    sum = 0
    for i in range(len(d)):
        for j in range(len(d[i])):
            sum += d[i][j]
    return sum / (len(d) * len(d[i]))

# DO NOT TOUCH THIS LINE
validate_fonction_caracteristique()
```

## Classification function

!!! tip Classification function  
We will now code the `classification` function which takes as input the feature calculated in the previous step and a threshold $t$, and returns the prediction of the algorithm.  
  
!!!
!!! question Complete the function code  
The function `classification` has the following parameters:  
- `x`: the feature of the image  
- `t`: the classification threshold  
and it must return 2 or 7, the prediction of the algorithm based on $x$ and $t$.  
!!!

```python
def classification(x, t):
    if x <= t:
        return 7
    else:
        return 2

# DO NOT TOUCH THIS LINE
validate_fonction_classification()
```

## Complete algorithm

!!! tip Algorithm function  
Using the functions `feature` and `classification`, we can code the complete algorithm which takes as input an image and returns the prediction 2 or 7.  
  
!!!
!!! question Complete the function code  
The function `algorithm` has the following parameter:  
- `d`: the image to classify  
and it must return 2 or 7, the prediction $\hat r$ of the algorithm based on the image $d$.  
!!!

```python
t = 30

def algorithm(d):
    return classification(feature(d), t)

# DO NOT TOUCH THIS LINE
validate_fonction_algorithme()
```

## Error computing

!!! info The percentage error: a way to evaluate your algorithm  
The lower the percentage error, the better is your solution.  
  
$$\text{percentage error} = \frac{\text{Number of misclassified images}} {\text{Total number of images}} \times 100$$  
A misclassified image is one for which the algorithm gave a prediction different from the true class of the image:  
$$\hat r \neq r$$  
!!!
!!! question Code the function to compute the percentage error  
Complete the code of the `compute_error` function.  
The first part of the function is already coded; it builds the list `r_estimations` of the responses 2 or 7 of your algorithm for each image. This list should be compared with `r_train` to calculate the percentage error.  
  
The function `compute_errror` takes as parameters:  
- `d_train`: A list of images  
- `r_train`: The list of true responses for each image in `d_train`  
It must return the percentage error of your algorithm.  
!!!

```python
def compute_error(d_train, r_train):
    r_estimations = []
    for i in range(len(d_train)):
        d = d_train[i] 
        estimation = algorithm(d)
        r_estimations.append(estimation)
    
    nb_errors = 0
    for i in range(len(r_train)):
        if r_train[i] != r_estimations[i]:
            nb_errors = nb_errors + 1

    erreur = nb_errors / len(d_train) * 100
    return erreur

# DO NOT TOUCH THIS LINE
validate_fonction_calcul_score()
```
